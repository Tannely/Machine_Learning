```python
import numpy as np
```

```python
'''1.	Створіть одновимірний масив (вектор) зі збільшувальною арифметичною послідовністю,
де кожне наступне число більше попереднього на 5, починаючи з 3, та виведіть його значення.'''

a = np.arange(3, 100, 5)
print(a)
```

```python
'''2.	Створіть двовимірний масив (матрицю) розміром 3x3, 
заповніть його добутками кожного числа та його попереднього, починаючи з 1, та виведіть його значення.'''

a = np.zeros((3, 3), dtype=int)

# Заповнюємо матрицю добутками чисел та їх попередніх
for i in range(3):
    for j in range(3):
        value = (i + 1) * (j + 1)
        a[i][j] = value
print(a)
```

```python
'''3.	Створіть масив розміром 8x8, заповніть його випадковими натуральними числами від 1 до 100 
та знайдіть середнє арифметичне його елементів.'''

a = np.random.randint(1, 101, (8, 8))

b = np.mean(a)

print("Масив:\n", a)
print("\nСереднє арифметичне = ", round(b,2))
```

```python
'''4.	Створіть масив розміром 4x4, заповніть його випадковими дійсними числами в діапазоні від 0 до 1 
та знайдіть кількість елементів, що більше 0.5.'''

a = np.random.rand(4, 4)
k = np.sum(a > 0.5)
print("Масив:\n", a)
print("\nКількість елементів, що більше 0.5 = ", k)
```

```python
'''5.	Створіть два одновимірних масиви розміром 7, 
заповніть їх випадковими цілими числами від -15 до 15 
та виконайте на них поелементні операції додавання, віднімання та множення.'''

a = np.random.randint(-15, 16, size=7)
b = np.random.randint(-15, 16, size=7)

print("a = ", a)
print("b = ", b)

print("a + b = ", a + b)
print("a - b = ", a - b)
print("a * b = ", a * b)
```

```python
'''6.	Створіть два вектори розміром 7, заповніть їх довільними цілими числами та знайдіть їх скалярний добуток.'''
a = np.random.randint(1, 11, size=7)
b = np.random.randint(3, 30, size=7)

dot_a_b = np.dot(a, b)

print("a = ", a)
print("b = ", b)
print("Скалярний добуток векторів = ", dot_a_b)
```

```python
'''7.	Створіть дві матриці розміром 4x2 та 2x5, 
заповніть їх випадковими цілими числами в діапазоні від 1 до 12 та перемножте їх між собою.'''

a = np.random.randint(1, 13, size=(4, 2))
b = np.random.randint(1, 13, size=(2, 5))

dot_a_b = np.dot(a, b)

print("a = ", a)
print("b = ", b)
print("Результат множення матриць:")
print(dot_a_b)
```

```python
'''8.	Створіть матрицю розміром 2x2, 
заповніть її випадковими цілими числами в діапазоні від 1 до 12 та знайдіть її обернену матрицю.'''

a = np.random.randint(1, 13, size=(2, 2))
print("Вихідна матриця:")
print(a)

# перевірка, чи можна знайти обернену матрицю
if np.linalg.det(a) == 0:
    print("Матриця не має оберненої матриці.")
else:
    # знаходження оберненої матриці
    inv_a = np.linalg.inv(a)
    print("Обернена матриця:")
    print(inv_a)
```

```python
'''9.	Створіть матрицю розміром 3x3,
заповніть її випадковими дійсними числами в діапазоні від 0 до 1 та транспонуйте її.'''

a = np.random.rand(3, 3)
print(a)

a_transp = np.transpose(a)
print("Транспонована матриця:")
print(a_transp)
```

```python
'''10.	Створіть матрицю розміром 3x4 та вектор розміром 4, заповніть матрицю випадковими дійсними числами від 0 до 1, 
а вектор - випадковими дійсними числами від 0 до 1, і знайдіть добуток вектора на матрицю.'''

matrix = np.random.rand(3, 4)
vector = np.random.rand(4)
# Знаходження добутку вектора на матрицю
m = matrix.reshape((4,3)) # зміна розмірності
result = np.dot(vector, m)  
print("Матриця:\n", matrix)
print("\nВектор:\n", vector)
print("\nРезультат добутку вектора на матрицю:\n", result)
```

```python
'''11.	Створіть матрицю розміром 2x3 та вектор розміром 3, 
заповніть матрицю випадковими дійсними числами від 0 до 1, 
а вектор - випадковими цілими числами в діапазоні від 1 до 10, і знайдіть добуток вектора на матрицю.'''

matrix = np.random.rand(2, 3)
vector = np.random.randint(1, 11, size=3)
# Знаходження добутку вектора на матрицю
m = matrix.reshape((3,2)) # зміна розмірності
result = np.dot(vector, m)  
print("Матриця:\n", matrix)
print("\nВектор:\n", vector)
print("\nРезультат добутку вектора на матрицю:\n", result)
```

```python
'''12.	Створіть дві матриці розміром 4x4, 
заповніть їх випадковими цілими числами в діапазоні від 1 до 10 та знайдіть різницю всіх елементів обох матриць.'''

a = np.random.randint(1, 11, size=(4, 4))
b = np.random.randint(1, 11, size=(4, 4))
print("Перша матриця:")
print(a)
print("Друга матриця:")
print(b)
print("Різниця всіх елементів обох матриць,\n", a-b)
```

```python
'''13.	Створіть дві матриці розміром 3x3, 
заповніть їх випадковими дійсними числами в діапазоні від -1 до 1 та знайдіть суму їх елементів.'''
matrix1 = np.random.randint(-1, 1, size=(3, 3))
matrix2 = np.random.randint(-1, 1, size=(3, 3))

# Знаходження суми елементів обох матриць
result = np.sum(matrix1) + np.sum(matrix2)

print("Перша матриця:")
print(matrix1)
print("\nДруга матриця:")
print(matrix2)
print("\nСума елементів матриць:")
print(result)
```

```python
'''14.	Створіть матрицю розміром 9x9, 
заповніть її випадковими дійсними числами в діапазоні від -10 до 10 та знайдіть кількість парних елементів матриці.'''

# Створення матриці 9x9 з випадковими дійсними числами в діапазоні від -10 до 10
matrix = np.random.randint(-10, 10, size=(9, 9))

# Знаходження кількості парних елементів у матриці
par = np.sum(matrix % 2 == 0)

print("Матриця 9x9:")
print(matrix)
print("\nКількість парних елементів у матриці = ", par)
```

```python
'''15.	Створіть матрицю розміром 5x5, заповніть її випадковими дійсними числами в діапазоні від -0.5 до 0.5 
та знайдіть вектор-стовпчик, що містить середнє геометричне значення елементів кожного рядка матриці.'''
matrix = np.random.uniform(-0.5, 0.5, size=(5, 5))
# Знаходимо середнє геометричне значення елементів кожного рядка
geom_means = np.prod(matrix, axis=1) ** (1/5)
geom_means_rounded = np.round(geom_means, 3)
geometric_means_column_vector = geom_means_rounded.reshape(-1, 1)

print("Матриця:")
print(matrix)
print("\nСереднє геометричне значення елементів кожного рядка:")
print(geom_means_rounded)
print("\nВектор-стовпчик з округленими середніми геометричними:")
print(geometric_means_column_vector)
```
